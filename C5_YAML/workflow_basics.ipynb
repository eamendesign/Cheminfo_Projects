```python
import yaml

with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)

print(config)
```

```python
polymer = config["system"]["polymer"]
T = config["simulation"]["temperature"]

print(f"Running {polymer} simulation at {T} K")
```

```python
new_config = {
    "system": {"polymer": "PTFE", "chain_length": 100},
    "simulation": {"temperature": 500}
}

with open("new_config.yaml", "w") as f:
    yaml.dump(new_config, f)
```
