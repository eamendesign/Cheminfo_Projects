```python
from ase import Atoms
from ase.visualize import view

# 创建水分子
water = Atoms(
    symbols=['O', 'H', 'H'],
    positions=[
        (0, 0, 0),
        (0.96, 0, 0),
        (-0.24, 0.93, 0)
    ]
)

print(water)

# 可视化
view(water)
```

```python
from ase.build import molecule

# 创建甲烷
ch4 = molecule('CH4')

print(ch4)

view(ch4)
```

```python
from ase.build import bulk
from ase.visualize import view

# 构建 FCC Cu
cu = bulk('Cu', 'fcc', a=3.6)

print(cu)

view(cu)
```

```python
from ase.build import fcc111
from ase.visualize import view

# 构建 Cu(111) 表面
slab = fcc111('Cu', size=(3,3,3), vacuum=10)

print(slab)

view(slab)
```

```python
from ase.build import bulk
from ase.visualize import view

cu = bulk('Cu', 'fcc', a=3.6)

# 构建 3x3x3 超晶胞
supercell = cu.repeat((3,3,3))

print(len(supercell), "atoms")

view(supercell)
```

```python
from ase.build import bulk
from ase.calculators.emt import EMT

cu = bulk('Cu', 'fcc', a=3.6)

# 设置计算器
cu.calc = EMT()

# 计算能量
energy = cu.get_potential_energy()

print("Energy =", energy, "eV")
```

```python
from ase.build import molecule
from ase.calculators.emt import EMT
from ase.optimize import BFGS

atoms = molecule("H2")

atoms.calc = EMT()

opt = BFGS(atoms)
opt.run(fmax=0.01)

print("Optimized positions:")
print(atoms.positions)
```

```python
from ase.build import molecule
from ase.io import write

atoms = molecule("H2O")

# 手动设置晶格（立方盒子，边长10Å）
atoms.set_cell([10.0, 10.0, 10.0])

# 将分子放在盒子中心
atoms.center()

# 保存为 xyz
write("water.xyz", atoms)

# 保存为 VASP POSCAR
write("POSCAR", atoms)
```

```python
from ase.build import molecule
from ase.calculators.emt import EMT
from ase.md.verlet import VelocityVerlet
from ase import units

atoms = molecule("H2")
atoms.calc = EMT()

dyn = VelocityVerlet(atoms, 1 * units.fs)

for i in range(10):
    dyn.run(1)
    print("Step", i, atoms.get_potential_energy())
```
