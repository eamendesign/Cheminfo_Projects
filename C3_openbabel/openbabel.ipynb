```python
from openbabel import pybel

# SMILES 表示的乙醇
smiles = "CCO"

# 读取 SMILES
mol = pybel.readstring("smi", smiles)

# 转换为 xyz 格式
xyz_string = mol.write("xyz")

print(xyz_string)
```

```python
smiles = "CCO"  # ethanol

mol = pybel.readstring("smi", smiles)

# 生成3D结构
mol.make3D()

# 保存为XYZ文件
mol.write("xyz", "ethanol.xyz", overwrite=True)

print("3D structure saved!")
```

```python
mol = pybel.readstring("smi", "CCO")

# 生成3D结构
mol.make3D()

# 使用UFF力场优化
mol.localopt(forcefield="uff", steps=200)

# 保存
mol.write("xyz", "ethanol_opt.xyz", overwrite=True)

print("Optimization finished")
```

```python
for mol in pybel.readfile("xyz", "ethanol.xyz"):
    
    name = mol.title
    
    mol.write("pdb", f"{name}.pdb", overwrite=True)
    
    print("converted:", name)
```

```python
mol = pybel.readstring("smi", "CCO")

print("Molecular weight:", mol.molwt)
print("Number of atoms:", len(mol.atoms))

# 计算分子指纹
fp = mol.calcfp()

print("Fingerprint bits:", len(fp.bits))
```

```python
smiles_list = [
    "CCO",
    "CCN",
    "CCC",
    "CCCl"
]

for i, smi in enumerate(smiles_list):

    mol = pybel.readstring("smi", smi)

    mol.make3D()
    mol.localopt()

    mol.write("xyz", f"mol_{i}.xyz", overwrite=True)

print("All molecules generated")
```

```python
mol = pybel.readstring("smi", "c1ccccc1")

print("Atoms:")

for atom in mol:
    print(atom.atomicnum, atom.coords)
```
